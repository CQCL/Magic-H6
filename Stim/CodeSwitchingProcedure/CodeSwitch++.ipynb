```python
import stim
```

```python
def bin_convert(list):
    list2 = []
    for l in list:
        if(l):
            list2.append(1)
        else:
            list2.append(0)
    return list2
```

```python
def checks(shot):
    #First, we propagate through the corrections from the Z/X measurements
    if (shot[2]):
        shot[4] = not shot[4]
    if (shot[3]):
        shot[8] = not shot[8]
        shot[9] = not shot[9]
        shot[6] = not shot[6]
    if (shot[0] or shot[1] or shot[4] or shot[5] or shot[6] or shot[7] or (shot[8]^shot[9]^shot[10]^shot[11])):
        return False
    return True
def success(shot):
    return not ((shot[8]^shot[11]) or (shot[8]^shot[10]))
```

```python
probs = [0,.001, .002,.003, .004, .005, .006]
logical_fails = []
for p in probs:
    start = 0
    anc = 6
    c = stim.Circuit()
    c.append("H", [start, start+2])
    c.append("CNOT", [start, anc])
    c.append("Depolarize2", [start, anc], p)
    c.append("CNOT", [start, start+1])
    c.append("Depolarize2", [start,start+1], p)
    c.append("CNOT", [start, start + 4])
    c.append("Depolarize2", [start,start+4], p)
    c.append("CNOT", [start, start + 5])
    c.append("Depolarize2", [start,start+5], p)
    c.append("CNOT", [start, anc])
    c.append("Depolarize2", [start, anc], p)
    c.append("CNOT", [start+2, anc+1])
    c.append("Depolarize2", [start+2,anc+1], p)
    c.append("CNOT", [start+2, start+3])
    c.append("Depolarize2", [start+2,start+3], p)
    c.append("CNOT", [start+2, start+4])
    c.append("Depolarize2", [start+2,start+4], p)
    c.append("CNOT", [start+2, start+5])
    c.append("Depolarize2", [start+2,start+5], p)
    c.append("CNOT", [start+2, anc+1])
    c.append("Depolarize2", [start+2,anc+1], p)
    c.append("MR", [anc, anc+1], p)
    c.append("H", [0,1,2,3,4,5])
    # We've fault-tolerantly initialized the codeblock to |++>_L in the [[6,2,2]] code
    
    
    c.append("H", [7]) # Storing the value of the [[6,2,2]] logical Z0 operator
    c.append("CX", [7,6])
    c.append("Depolarize2", [6, 7], p)
    
    c.append("CX", [0,6])
    c.append("Depolarize2", [6, 0], p)
    c.append("CX", [3,7])
    c.append("Depolarize2", [7, 3], p)
    c.append("CX", [5,6])
    c.append("Depolarize2", [6, 5], p)

    c.append("H", [8]) # Storing the value of the [[6,2,2]] logical X1 operator
    c.append("CX", [8,9])
    c.append("Depolarize2", [8, 9], p)
    
    c.append("CX", [8,1])
    c.append("Depolarize2", [8, 1], p)
    c.append("CX", [9,2])
    c.append("Depolarize2", [9, 2], p)
    c.append("CX", [8,4])
    c.append("Depolarize2", [8, 4], p)

    # We measure the 4th and last qubit in the Z/X basis to go to [[4,2,2]]
    c.append("H", [5])
    c.append("MR", [4,5], p)

    #We remeasure the logical Z0 operator now for [[4,2,2]]. We have applied the logical Xs based on the Z_4 measurement in post-processing
    c.append("CX", [0,6])
    c.append("Depolarize2", [6, 0], p)
    c.append("CX", [2,7])
    c.append("Depolarize2", [7, 2], p)

    #We remeasure the logical X1 operator now for [[4,2,2]]. We have applied the logical Zs based on the X_5 measurement in post-processing
    c.append("CX", [8,1])
    c.append("Depolarize2", [8, 1], p)
    c.append("CX", [9,3])
    c.append("Depolarize2", [9, 3], p)

    # Now we measure the ancillae in the appropriate bases
    c.append("CX", [7,6])
    c.append("CX", [8, 9])
    c.append("H", [7,8])
    c.append("MR", [6,7,8,9], p)

    # Now we measure the codeblock in the X basis to see if the state is |++>_L in the [[4,2,2]] code
    c.append("H", [0,1,2,3])
    c.append("MR", [0,1,2,3], p)
    
    s = c.compile_sampler()
    l = s.sample(10000000)
    fails = 0
    totals = 0
    for shot in l:
        if (checks(shot)):
            totals += 1
            if (not success(shot)):
                fails += 1
    logical_fails.append(fails/totals)
```

```python
print(fails,totals)
print(logical_fails)
```

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.optimize import curve_fit
def func(x, a, b):
    return a * x**b
```

```python
#This is the undistilled function
print(curve_fit(func, probs, logical_fails))
predicted = []
for p in probs:
    predicted.append(func(p, 20.94456695, 2.05245848))
```

```python
plt.plot(probs, logical_fails, 'bo', label = 'Logical Failure Rate of Code Switching')
plt.plot(probs, predicted, label = 'Predicted')
plt.xlabel('Physical Error Rate')
plt.ylabel('[[4,2,2]] [[4,2,2]] |++>_L State Infidelity')
plt.title('Logical Failure Rate vs TQ Error Probability')
plt.legend()
```
